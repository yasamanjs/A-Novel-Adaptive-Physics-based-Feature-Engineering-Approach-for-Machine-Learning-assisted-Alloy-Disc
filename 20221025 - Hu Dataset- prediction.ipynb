```python
############################## 1- Read Data ####################################

#### Pandas is used for data manipulation
import pandas as pd
import numpy as np


# features = pd.read_excel('Data_Hu_original.xlsx')
features = pd.read_excel('Data_Hu_original-AE.xlsx', sheet_name='Original+AE')
# features = pd.read_excel('Data_Hu_original-AE.xlsx', sheet_name='AE+clean')


features = features.drop('Ga', axis=1)
features = features.drop('Bi', axis=1)
features = features.drop('Pb', axis=1)
features = features.drop('B', axis=1)
features = features.drop('Be', axis=1)
features = features.drop('Ni', axis=1)



features = features.drop('Reference', axis=1)
features = features.drop('Alloy', axis=1)

# features = features.drop('Designation', axis=1)
features = features.drop('Temper_type', axis=1)
features = features.drop('Temper_digit1', axis=1)
features = features.drop('Temper_digit2', axis=1)


features = features.drop('S/A Temp', axis=1)
features = features.drop('Ageing Temp', axis=1)
features = features.drop('Ageing Time', axis=1)
features = features.drop('Manufacturing Process', axis=1)
features = features.drop('Production Shape', axis=1)
features = features.drop('Heat Treatment Medium', axis=1)
features = features.drop('Ageing Type', axis=1)
features = features.drop('Treatment Afterward', axis=1)


# features = features.drop('Strain-hardening Method', axis=1)
features = features.drop('SHM', axis=1)
features = features.drop('SHM-level', axis=1)
```

```python
############   keep one properety / remove missing
features = features.drop('UTS', axis=1)
features = features.drop('YTS', axis=1)
# features = features.drop('ELONGATION', axis=1)

features = features.dropna()
features
```

```python
### defining label
# label1 = np.array(features['UTS'])
# label1 = np.array(features['YTS'])
label1 = np.array(features['ELONGATION'])


# features= features.drop('UTS', axis = 1)
# features= features.drop('YTS', axis = 1)
features= features.drop('ELONGATION', axis = 1)



n_samples, n_features = features.shape
# One-hot encode categorical features (to encode)
features = pd.get_dummies(features)

# Saving feature names for later use
feature_list = list(features.columns)
features = np.array(features)
```

```python
############################# evaluate function #############################
from sklearn.metrics import mean_absolute_error

def evaluate(true_labels, pred_labels):
    errors = abs(pred_labels - true_labels)
    MAE = round(np.mean(errors), 2)
    MAE2 = mean_absolute_error(true_labels, pred_labels)
    mape = 100 * np.mean(errors / true_labels)
    #R2, and adjusted R2
    from sklearn.metrics import r2_score
    r2 = r2_score(true_labels, pred_labels)

    ## p is the total number of explanatory variables in the model
    ## and n is the sample size.
    p = n_features
    n = n_samples
    
    r2_adjusted = 1-(1-r2)*(n-1)/(n-p-1)
    r2_adjusted = np.array(r2_adjusted)


    from sklearn.metrics import mean_squared_error
    RMSE = mean_squared_error(true_labels, pred_labels,squared=False)

    #return accuracy, MAE, MAPE, r2_adjusted, r2_mean_5cv
    return mape, MAE, RMSE, r2, r2_adjusted
```

```python
  ################### the loop for trying 100 seeds ###########################
  ##### at each iteration, a random seed is defined, 
  ##### test/train is split for that seed, models are devloped, 
  ##### trained and evaluated.
from sklearn.model_selection import train_test_split
i = 0

for j in range(100): #trying range of 0 to 99 for seeds to do the split
# for j in A:
    Seed = j
    print(j)
    
    test_ratio = 0.2

    train_features1, test_features1, train_labels1, test_labels1 = train_test_split(features, label1,
                                                                                test_size = test_ratio,
                                                                                random_state = Seed)

    print(i)
    ################## build Model ######################################
    ################################################################################
    SEED = 100
    # ######################## Random Forest ###################################
    from sklearn.ensemble import RandomForestRegressor
    ## RF default
    Model1 = RandomForestRegressor(random_state=SEED)

    ################# Evaluation ##################

    Model1.fit(train_features1,train_labels1)

    print('###################################################################')
    print('evaluatoin of the RF model on testing dataset')

    predictions1 = Model1.predict(test_features1)


    model_accuracy1, model_MAE1, model_RMSE1, model_r21, model_r2_adjusted1 = evaluate(test_labels1, predictions1)


    final_output.loc[:,i] = [model_accuracy1, model_MAE1, model_RMSE1, model_r21, model_r2_adjusted1]


    i = i+1
    i

```

```python
### collecting evaluation metrices for all 100 seeds for each property
final_output = pd.DataFrame(final_output)
final_output = final_output.T

Final_results_mean = [final_output.mean()]
Final_results_std = [final_output.std()]

Final_results_mean = pd.DataFrame(Final_results_mean)
Final_results_std = pd.DataFrame(Final_results_std)
```
