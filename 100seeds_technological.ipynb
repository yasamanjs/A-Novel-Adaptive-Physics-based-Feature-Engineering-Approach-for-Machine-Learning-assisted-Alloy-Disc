```python
############################## 1- Read Data ####################################

# Pandas is used for data manipulation
import pandas as pd
import numpy as np


# upload data
# features = pd.read_excel('Alldata236.xlsx')
features = pd.read_excel('Alldata236_ordinal.xlsx')
# features = pd.read_excel('Alldata236_12types.xlsx')
# features = pd.read_excel('Alldata236_AE.xlsx', sheet_name='4tempercolumn')
```

```python
features
```

```python
################# defining labels: technological properties

features = features.replace('Borderline [3]', 'Borderline')
features = features.replace('Fair [3]', 'Fair')
features = features.replace('Excellent [3]', 'Excellent')
features = features.replace('Very Good [3]', 'Very Good')
features = features.replace('Unacceptable or Not Commericially Practical', 'Unacceptable')

# Labels are the values we want to predict
label1 = np.array(features['RGC'])
label2 = np.array(features['RSCC'])
label3 = np.array(features['Extr'])
label4 = np.array(features['CW'])
label5 = np.array(features['Mach'])
label6 = np.array(features['RSW'])
label7 = np.array(features['Braz'])
label8 = np.array(features['GW'])
label9 = np.array(features['AW'])
```

```python
#### Remove the labels, properties, and extra columns from the features
#### axis 1 refers to the columns
features= features.drop('UTS', axis = 1)
features= features.drop('TYS', axis = 1)
features= features.drop('El', axis = 1)
features= features.drop('BHN', axis = 1)
features= features.drop('Shear', axis = 1)
features= features.drop('Endu', axis = 1)

features= features.drop('alloy name', axis = 1)


features= features.drop('RGC', axis = 1)
features= features.drop('RSCC', axis = 1)
features= features.drop('CW', axis = 1)
features= features.drop('Mach', axis = 1)
features= features.drop('Braz', axis = 1)
features= features.drop('GW', axis = 1)
features= features.drop('AW', axis = 1)
features= features.drop('RSW', axis = 1)
features= features.drop('Extr', axis = 1)

features= features.drop('temper_type', axis = 1)
features= features.drop('encoded_temper', axis = 1)

n_samples, n_features = features.shape
# One-hot encode categorical features (Encoding tempers)
features = pd.get_dummies(features)
features.head(5)

# Saving feature names for later use
feature_list = list(features.columns)

# Convert to numpy array
features = np.array(features)
```

```python
features.shape
```

```python
##################### evaluation ###############################################
target_names = ['Unacceptable', 'Borderline','Fair', 'Very Good', 'Excellent']

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

def evaluate(test_labels, predictions):
    cm = confusion_matrix(test_labels, predictions, labels=target_names)
    Report = classification_report(test_labels, predictions, labels=target_names, output_dict=True)
    Report = pd.DataFrame(Report)
    avg = Report['weighted avg']
    Conf_Mat = pd.DataFrame(cm)
    Report = Report.append(Conf_Mat)
    return Report, avg, cm
```

```python
################### the loop for 100 seeds       ###############################
#### at each iteration, seed is defined, test/trainis split, 
### models built, trained and evaluated.
from sklearn.model_selection import train_test_split

final_output1 = pd.DataFrame()
final_output2 = pd.DataFrame()
final_output3 = pd.DataFrame()
final_output4 = pd.DataFrame()
final_output5 = pd.DataFrame()
final_output6 = pd.DataFrame()
final_output7 = pd.DataFrame()
final_output8 = pd.DataFrame()
final_output9 = pd.DataFrame()

i = 0

for j in range(100):
    i = i + 1
    SEED = j

    train_features1, test_features1, train_labels1, test_labels1 = train_test_split(features, label1,
                                                                                      test_size = 0.2,
                                                                                      random_state = SEED,
                                                                                      stratify=label1)


    train_features2, test_features2, train_labels2, test_labels2 = train_test_split(features, label2,
                                                                                  test_size = 0.2,
                                                                                  random_state = SEED,
                                                                                      stratify=label2)


    train_features3, test_features3, train_labels3, test_labels3 = train_test_split(features, label3,
                                                                                  test_size = 0.2,
                                                                                  random_state = SEED,
                                                                                      stratify=label3)


    train_features4, test_features4, train_labels4, test_labels4 = train_test_split(features, label4,
                                                                                  test_size = 0.2,
                                                                                  random_state = SEED,
                                                                                      stratify=label4)


    train_features5, test_features5, train_labels5, test_labels5 = train_test_split(features, label5,
                                                                                  test_size = 0.2,
                                                                                  random_state = SEED,
                                                                                      stratify=label5)


    train_features6, test_features6, train_labels6, test_labels6 = train_test_split(features, label6,
                                                                                  test_size = 0.2,
                                                                                  random_state = SEED,
                                                                                      stratify=label6)


    train_features7, test_features7, train_labels7, test_labels7 = train_test_split(features, label7,
                                                                                  test_size = 0.2,
                                                                                  random_state = SEED,
                                                                                      stratify=label7)

    train_features8, test_features8, train_labels8, test_labels8 = train_test_split(features, label8,
                                                                                  test_size = 0.2,
                                                                                  random_state = SEED,
                                                                                      stratify=label8)
    train_features9, test_features9, train_labels9, test_labels9 = train_test_split(features, label9,
                                                                                  test_size = 0.2,
                                                                                  random_state = SEED,
                                                                                      stratify=label9)



    #### RF
    from sklearn.ensemble import RandomForestClassifier
    SEED = 100
    #### RF default
    Model1 = RandomForestClassifier(random_state=SEED)
    Model2 = RandomForestClassifier(random_state=SEED)
    Model3 = RandomForestClassifier(random_state=SEED)
    Model4 = RandomForestClassifier(random_state=SEED)
    Model5 = RandomForestClassifier(random_state=SEED)
    Model6 = RandomForestClassifier(random_state=SEED)
    Model7 = RandomForestClassifier(random_state=SEED)
    Model8 = RandomForestClassifier(random_state=SEED)
    Model9 = RandomForestClassifier(random_state=SEED)



    ################# Evaluation ##################

    Model1.fit(train_features1,train_labels1)
    Model2.fit(train_features2,train_labels2)
    Model3.fit(train_features3,train_labels3)
    Model4.fit(train_features4,train_labels4)
    Model5.fit(train_features5,train_labels5)
    Model6.fit(train_features6,train_labels6)
    Model7.fit(train_features7,train_labels7)
    Model8.fit(train_features8,train_labels8)
    Model9.fit(train_features9,train_labels9)
    print('###################################################################')
    print('evaluatoin of the RF model on testing dataset')

    predictions1 = Model1.predict(test_features1)
    predictions2 = Model2.predict(test_features2)
    predictions3 = Model3.predict(test_features3)
    predictions4 = Model4.predict(test_features4)
    predictions5 = Model5.predict(test_features5)
    predictions6 = Model6.predict(test_features6)
    predictions7 = Model7.predict(test_features7)
    predictions8 = Model8.predict(test_features8)
    predictions9 = Model9.predict(test_features9)

    report1, avg1, cm1  = evaluate(test_labels1, predictions1)
    report2, avg2, cm2  = evaluate(test_labels2, predictions2)
    report3, avg3, cm3  = evaluate(test_labels3, predictions3)
    report4, avg4, cm4  = evaluate(test_labels4, predictions4)
    report5, avg5, cm5  = evaluate(test_labels5, predictions5)
    report6, avg6, cm6  = evaluate(test_labels6, predictions6)
    report7, avg7, cm7  = evaluate(test_labels7, predictions7)
    report8, avg8, cm8  = evaluate(test_labels8, predictions8)
    report9, avg9, cm9  = evaluate(test_labels9, predictions9)


    final_output1.loc[:,i] = avg1
    final_output2.loc[:,i] = avg2
    final_output3.loc[:,i] = avg3
    final_output4.loc[:,i] = avg4
    final_output5.loc[:,i] = avg5
    final_output6.loc[:,i] = avg6
    final_output7.loc[:,i] = avg7
    final_output8.loc[:,i] = avg8
    final_output9.loc[:,i] = avg9

    print(i)
```

```python
######## 100 evaluation metrices for each property
final_output1 = final_output1.T
final_output2 = final_output2.T
final_output3 = final_output3.T
final_output4 = final_output4.T
final_output5 = final_output5.T
final_output6 = final_output6.T
final_output7 = final_output7.T
final_output8 = final_output8.T
final_output9 = final_output9.T
```
